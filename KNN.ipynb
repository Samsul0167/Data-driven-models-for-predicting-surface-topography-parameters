```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, make_scorer
from sklearn.inspection import permutation_importance
import joblib

# Load the data from a CSV file
data = pd.read_csv('PST_v6.csv')  # Replace 'topography.csv' with the actual file name

# Remove commas and convert to numeric values
data = data.replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')
data = data.dropna()

# Define input (X) and output (y) variables
X = data[['cpower1', 'cspeed1', 'cpower2', 'cspeed2', 'angle']]
y = data[['eta_SID', 'rho_s_SID', 'sig_s_SID', 'Sa', 'Sq', 'Ssk', 'Sku', 'sigma_s_rho', 'Sdq', 'Sdr']]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Scale the input features for better performance with KNN
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the hyperparameters for grid search for KNN
param_grid_knn = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'p': [1, 2]  # 1 for Manhattan distance, 2 for Euclidean distance
}

# Dictionary to store best hyperparameters, R², and other metrics for each target
best_params_dict_knn = {}
metrics_dict_knn = {}

# Directory to save the models
model_save_path = 'knn_models/'

# Create the directory if it doesn't exist
if not os.path.exists(model_save_path):
    os.makedirs(model_save_path)

# Initialize an empty DataFrame to store actual and predicted values
results_df = pd.DataFrame()

# Function to sanitize filenames
def sanitize_filename(name):
    return name.replace('\\', '_').replace('/', '_')

# Loop through each target variable to train and save the model and hyperparameters together
for i, col in enumerate(y.columns):
    print(f"Optimizing KNN for target variable: {col}")
    
    # Perform Grid Search with cross-validation for each target
    grid_search = GridSearchCV(estimator=KNeighborsRegressor(), param_grid=param_grid_knn, cv=10, n_jobs=-1, verbose=2)
    grid_search.fit(X_train_scaled, y_train.iloc[:, i])

    # Get the best parameters for the current target
    best_params = grid_search.best_params_
    best_params_dict_knn[col] = best_params
    print(f"Best Parameters for {col}: {best_params}")
    
    # Get the best KNN estimator for the current target
    best_knn = grid_search.best_estimator_

    # Perform cross-validation on the training set
    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    r2_scores = cross_val_score(best_knn, X_train_scaled, y_train.iloc[:, i], cv=kf, scoring='r2')
    rmse_scores = cross_val_score(best_knn, X_train_scaled, y_train.iloc[:, i], cv=kf, scoring=make_scorer(mean_squared_error, squared=False))

    print(f"{col} - Cross-Validated R² Scores: {r2_scores}")
    print(f"{col} - Cross-Validated RMSE Scores: {rmse_scores}")

    # Make predictions on the test set
    y_pred = best_knn.predict(X_test_scaled)

    # Evaluate the model on the test set
    r2_test = r2_score(y_test.iloc[:, i], y_pred)
    rmse_test = np.sqrt(mean_squared_error(y_test.iloc[:, i], y_pred))
    mae_test = mean_absolute_error(y_test.iloc[:, i], y_pred)

    # Calculate normalized RMSE and normalized MAE
    range_target = y_test.iloc[:, i].max() - y_test.iloc[:, i].min()
    normalized_rmse = rmse_test / range_target
    normalized_mae = mae_test / range_target

    # Store metrics for later display
    metrics_dict_knn[col] = {
        'R² (Train Cross-Validated)': np.mean(r2_scores),
        'R² (Test)': r2_test,
        'RMSE (Test)': rmse_test,
        'Normalized RMSE (Test)': normalized_rmse,
        'MAE (Test)': mae_test,
        'Normalized MAE (Test)': normalized_mae
    }

    print(f"{col} - Test Set R² Score: {r2_test:.4f}, RMSE: {rmse_test:.4f}, Normalized RMSE: {normalized_rmse:.4f}, MAE: {mae_test:.4f}, Normalized MAE: {normalized_mae:.4f}")
    
    # Save the trained model for this target variable
    sanitized_col = sanitize_filename(col)
    model_filename = os.path.join(model_save_path, f'knn_model_{sanitized_col}.pkl')
    model_and_params = {'model': best_knn, 'best_params': best_params}
    joblib.dump(model_and_params, model_filename)
    print(f"Model and hyperparameters for {col} saved as {model_filename}")
    
    # Store the actual and predicted values for the current roughness parameter
    temp_df = pd.DataFrame({
        'Model': 'KNN',
        'Roughness Parameter': col,
        'Actual': y_test.iloc[:, i],
        'Predicted': y_pred
    })

    # Append the results to the main DataFrame
    results_df = pd.concat([results_df, temp_df], ignore_index=True)

    # Plotting Actual vs Predicted for each target variable
    plt.figure(figsize=(10, 6))
    plt.scatter(y_test.iloc[:, i], y_pred, color='blue')
    plt.plot([y_test.iloc[:, i].min(), y_test.iloc[:, i].max()], 
             [y_test.iloc[:, i].min(), y_test.iloc[:, i].max()], color='red', lw=2)
    plt.xlabel(f'Actual {col}')
    plt.ylabel(f'Predicted {col}')
    plt.title(f'Actual vs Predicted for {col}')
    plt.show()

    # Permutation Feature Importance (PFI) for each output
    pfi = permutation_importance(best_knn, X_test_scaled, y_test.iloc[:, i], n_repeats=10, random_state=42, n_jobs=-1)
    sorted_idx = pfi.importances_mean.argsort()
    
    plt.figure(figsize=(10, 6))
    plt.barh(X.columns[sorted_idx], pfi.importances_mean[sorted_idx])
    plt.xlabel("Permutation Feature Importance")
    plt.title(f"Feature Importance based on PFI for {col}")
    plt.show()

# Export the actual and predicted values to a CSV file
results_df.to_csv('actual_vs_predicted_results_knn.csv', index=False)
print("Actual vs Predicted results saved to 'actual_vs_predicted_results_knn.csv'")

# Optional: Save the best hyperparameters dictionary for all targets
joblib.dump(best_params_dict_knn, 'best_hyperparameters_per_target_knn.pkl')

# Display all best hyperparameters and test metrics
print("\nBest Hyperparameters and Metrics for Each Target Variable (KNN):")
for target, params in best_params_dict_knn.items():
    print(f"{target}: {params}")

# Display the test set R², RMSE, and MAE for each target
print("\nTest Set Metrics for Each Target Variable (KNN):")
for target, metrics in metrics_dict_knn.items():
    print(f"{target}: {metrics}")
```
