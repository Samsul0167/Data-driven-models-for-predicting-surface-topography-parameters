```python
# Necessary imports
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score, KFold
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, make_scorer
from sklearn.inspection import permutation_importance
import joblib

# Load the data from a CSV file
data = pd.read_csv('PST_v6.csv')  # Replace 'topography.csv' with the actual file name

# Remove commas and convert to numeric values
data = data.replace(',', '', regex=True).apply(pd.to_numeric, errors='coerce')
data = data.dropna()

# Define input (X) and output (y) variables
X = data[['cpower1', 'cspeed1', 'cpower2', 'cspeed2', 'angle']]
y = data[['eta_SID', 'rho_s_SID', 'sig_s_SID', 'Sa', 'Sq', 'Ssk', 'Sku', 'sigma_s_rho', 'Sdq', 'Sdr']]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=42)

# Scale the input features for better performance with SVM
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Define the hyperparameters for grid search for SVM
param_grid_svm = {
    'kernel': ['linear', 'rbf', 'poly'], 
    'C': [0.01,0.1, 1, 10],
    'epsilon': [0.01, 0.1, 1,10],
    'gamma': ['scale', 'auto']
}

# Dictionary to store best hyperparameters and metrics for each target
best_params_dict_svm = {}
metrics_dict_svm = {}

# Function to sanitize filenames
def sanitize_filename(name):
    return name.replace('\\', '_').replace('/', '_')

# Directory to save the models
model_save_path = 'svm_models/'

# Create the directory if it doesn't exist
if not os.path.exists(model_save_path):
    os.makedirs(model_save_path)

# Loop through each target variable to train and save the model
for i, col in enumerate(y.columns):
    print(f"Optimizing SVM for target variable: {col}")
    
    # Perform Grid Search with cross-validation for each target
    grid_search = GridSearchCV(estimator=SVR(), param_grid=param_grid_svm, cv=10, n_jobs=-1, verbose=2)
    grid_search.fit(X_train_scaled, y_train.iloc[:, i])

    # Get the best parameters for the current target
    best_params = grid_search.best_params_
    best_params_dict_svm[col] = best_params
    print(f"Best Parameters for {col}: {best_params}")
    
    # Train the SVM Regressor with the best parameters for the current target
    best_svm = grid_search.best_estimator_

    # Perform cross-validation
    kf = KFold(n_splits=10, shuffle=True, random_state=42)
    r2_scores = cross_val_score(best_svm, X_train_scaled, y_train.iloc[:, i], cv=kf, scoring='r2')
    rmse_scores = cross_val_score(best_svm, X_train_scaled, y_train.iloc[:, i], cv=kf, scoring=make_scorer(mean_squared_error, squared=False))

    print(f"{col} - Cross-Validated R2 Scores: {r2_scores}")
    print(f"{col} - Cross-Validated RMSE Scores: {rmse_scores}")

    # Make predictions on the test set
    y_pred = best_svm.predict(X_test_scaled)

    # Evaluate the model on the test set
    r2 = r2_score(y_test.iloc[:, i], y_pred)
    rmse = np.sqrt(mean_squared_error(y_test.iloc[:, i], y_pred))
    mae = mean_absolute_error(y_test.iloc[:, i], y_pred)

    # Calculate normalized RMSE and normalized MAE
    range_target = y_test.iloc[:, i].max() - y_test.iloc[:, i].min()
    normalized_rmse = rmse / range_target
    normalized_mae = mae / range_target

    metrics_dict_svm[col] = {
        'R2': r2, 
        'RMSE': rmse, 
        'Normalized RMSE': normalized_rmse, 
        'MAE': mae, 
        'Normalized MAE': normalized_mae
    }
    
    print(f"{col} - Test Set R2 Score: {r2:.4f}, RMSE: {rmse:.4f}, Normalized RMSE: {normalized_rmse:.4f}, MAE: {mae:.4f}, Normalized MAE: {normalized_mae:.4f}")
    
    # Save the trained model for this target variable
    sanitized_col = sanitize_filename(col)
    model_filename = os.path.join(model_save_path, f'svm_model_{sanitized_col}.pkl')
    joblib.dump(best_svm, model_filename)
    print(f"Model for {col} saved as {model_filename}")
    
    # Plotting individual Actual vs Predicted for each target variable
    plt.figure(figsize=(10, 6))
    plt.scatter(y_test.iloc[:, i], y_pred, color='blue')
    plt.plot([y_test.iloc[:, i].min(), y_test.iloc[:, i].max()], 
             [y_test.iloc[:, i].min(), y_test.iloc[:, i].max()], color='red', lw=2)
    plt.xlabel(f'Actual {col}')
    plt.ylabel(f'Predicted {col}')
    plt.title(f'Actual vs Predicted for {col}')
    plt.show()

    # Permutation Feature Importance (PFI) for each output
    pfi = permutation_importance(best_svm, X_test_scaled, y_test.iloc[:, i], n_repeats=10, random_state=42, n_jobs=-1)
    sorted_idx = pfi.importances_mean.argsort()
    
    plt.figure(figsize=(10, 6))
    plt.barh(X.columns[sorted_idx], pfi.importances_mean[sorted_idx])
    plt.xlabel("Permutation Feature Importance")
    plt.title(f"Feature Importance based on PFI for {col}")
    plt.show()

# Display all best hyperparameters for each target variable
print("\nBest Hyperparameters for Each Target Variable (SVM):")
for target, params in best_params_dict_svm.items():
    print(f"{target}: {params}")

# Correlation Heatmap
combined_data = pd.concat([X, y], axis=1)
corr_matrix = combined_data.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5, annot_kws={"size": 10}, fmt=".2f")
plt.xticks(rotation=45, ha='right',fontsize=18)
plt.yticks(rotation=0,fontsize=18)
plt.title('Correlation Heatmap', fontsize=18)
plt.tight_layout()
plt.show()

# Optional: Save the best hyperparameters dictionary
joblib.dump(best_params_dict_svm, 'best_hyperparameters_per_target_svm.pkl')
```
